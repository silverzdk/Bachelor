```python
import sympy as sp 
import sympy.physics.units as u

d_i, d_o, h_h, h_c = sp.symbols('d_i d_o h_h h_c')

U_I = (1/h_c + d_i/(d_o*h_h))**-1
display(U_I)
```

```python
m_dot = 0.1 * u.kg / u.s
inner_diameter = 0.05 * u.m
density = 1000 * u.kg / u.m**3

vel = m_dot / (density * (sp.pi * (inner_diameter / 2)**2))
display(vel.evalf(3))
```
