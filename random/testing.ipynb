#### Example of fluid class

```python
# Define the fluid (water in this case)
water = Fluid(FluidsList.Water)

# Set the state 
water.update(Input.enthalpy(430340.09), Input.pressure(101325))  # 1 bar = 100000 Pa

# Get the specific enthalpy in J/kg
enthalpy = water.H

print(f"Enthalpy {enthalpy/1000:.2f} kJ/kg")
print(f"Temperature {water.temperature } 째C")
print(f"Quality {water.quality }")

pipe = Pipe(length=10,
            outer_diameter=0.1,
            wall_thickness=0.005,
            angle=0)  # angle in degrees

water.set_geometry(pipe)
water.liquid_phase().H

water.m_dot = 0.5
water_100C = water.with_state(Input.quality(0.5), Input.temperature(100))

print(f"Enthalpy {water_100C.H/1000:.2f} kJ/kg")
print(f"Temperature {water_100C.temperature } 째C")
print(f"Quality {water_100C.quality }")

print(water.temperature)
```
